```python
import pandas as pd

fert = pd.read_csv('./data/gapminder_total_fertility.csv', index_col=0)
```

```python
life = pd.read_excel('./data/gapminder_lifeexpectancy.xlsx', index_col=0)
```

```python
ncol = [int(x) for x in fert.columns]
fert.set_axis(axis=1, labels=ncol, inplace=True)
```

```python
sfert = fert.stack()
```

```python
slife = life.stack()
```

```python
d = {'fertility': sfert, 'lifeexp': slife}
df2 = pd.DataFrame(data=d)
```

```python
df2
```

```python
popul = pd.read_excel('./data/gapminder_population.xlsx', index_col=0)
```

```python
print(popul.shape)
```

```python
popul.columns
```

```python
spopul=popul.stack()
```

```python
spopul
```

```python
d = {'fertility': sfert, 'lifeexp': slife, 'population': spopul}
df2 = pd.DataFrame(data=d)
df2
```

```python
df3 = df2.stack()
df3
```

```python
df4 = df3.unstack((0,2))
df4
```

```python
import pylab as plt
df4[['Germany', 'France', 'Sweden']].plot()
```

```python
df5 = df3.unstack(2)
df5.plot.scatter('fertility', 'lifeexp', s=0.1)
```

```python
df6 = df3.unstack(1)
df6 = df6[1950]
df6 = df6.unstack(1)
df6.plot.scatter('fertility', 'lifeexp', s=3)
```

```python
cmap = plt.get_cmap('tab20', lut = len(df6)).colors
df6.plot.scatter('fertility', 'lifeexp', s=10, c=cmap)
```

```python

cmap = plt.get_cmap("tab20", lut = len(df6)).colors
df6.plot.scatter('fertility', 'lifeexp', s=df6['population']/150000, c=cmap)
```

```python
y = 1960

for i in range(1960,2016):
    df6 = df3.unstack(1)
    df6 = df6[y]
    df6 = df6.unstack(1)
    df6.plot.scatter('fertility', 'lifeexp', s=df6['population']/100000, c=cmap)
    plt.axis((0, 10, 0, 120))
    plt.title(y)
    # get rid of the frame
    for spine in plt.gca().spines.values():
        spine.set_visible(False)
    plt.savefig(f"lifeexp_{y}.png")
    y = y+1
```
