```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
plt.rcParams['figure.figsize'] = (10,6)
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
```

### Understanding the data

- SOUID: Source identifier
- DATE : Date YYYYMMDD
- TG   : mean temperature in 0.1 &#176;C
- Q_TG : Quality code for TG (0='valid'; 1='suspect'; 9='missing')
- This is the blended series of station GERMANY, BERLIN-TEMPELHOF (STAID: 2759).
- Blended and updated with sources: 111448 127488 128124 

```python
df=pd.read_csv("./data/TG_STAID002759.txt", parse_dates=True, index_col=1)
df.rename(columns = {' SOUID': 'source', '   TG':"temp", ' Q_TG': 'QTG'}, inplace=True)
df['temp']=df['temp']*0.1
```

```python
missing = pd.date_range('1945-04-25', periods=195, freq='D')  # solution for loops from Aron
year44 = pd.date_range('1944-04-25', periods=195, freq='D')
year46 = pd.date_range('1946-04-25', periods=195, freq='D')
m=0
for i in tqdm(missing):
    df.loc[i,"temp"] = (df.loc[year44[m]]["temp"] + df.loc[year46[m]]["temp"]) / 2
    m += 1
```

```python
df['tempK']=df['temp']+273.15 #Kelvin
```

```python
df.head(2)
```

```python
df=df.drop(['QTG', 'source','temp'],axis=1) #removing quality code with 1=suspect and 9=missing
```

```python
df.plot() 
plt.xlabel('Years', fontsize=15)
plt.ylabel('Temperature', fontsize=15)
plt.title('Mean temperature in Berlin Templehof' )
```

```python
df.loc['2010':'2020'].plot()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Temperature', fontsize=15)
plt.title('Mean temperature in Berlin Templehof from 2010 to 2020' )
```

#### Differential: Trend

```python
df['diff']=df['tempK'].diff()  
df['diff'].plot()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Temperature difference', fontsize=15)
plt.title('Temperature difference in Berlin Templehof' )
```

```python
# how the temperature changes every year:
df.groupby(df.index.year)['diff'].mean()
```

```python
df.groupby(df.index.year)['diff'].mean().plot()
plt.xlabel('Years', fontsize=15)
plt.ylabel('Temperature difference', fontsize=15)
plt.title('Temperature change every year in Berlin Templehof' )
```

#### Percentage change
removes trend and variance change

```python
df['pct'] = df['tempK'].pct_change()
```

```python
df['pct'].plot()
```

```python
df.groupby([df.index.month])['tempK'].mean().plot()
```

```python
df.groupby([df.index.year])["tempK"].mean().plot()
```

### Seasonality

```python
df['month_mean']=df.groupby(df.index.month)['pct'].transform('mean')
```

```python
df['month_mean']
```

```python
df['month_mean'].plot()
```

```python
df['detrend_div'] = df['pct'] / df['month_mean']
```

```python
df['detrend_div'].mean()
```

```python
df['detrend_div'].hist(bins=10)
```

```python
df['detrend_div'].plot()
```

```python
df['detrend_sub'] = df['pct'] - df['month_mean']
```

```python
df['detrend_sub'].plot.hist(bins=20)
```

```python
df['detrend_sub'].plot()
```
