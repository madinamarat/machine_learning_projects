```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,6)
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df=pd.read_csv("./data/TG_STAID002759.txt", parse_dates=True, index_col=1)
df.rename(columns = {' SOUID': 'source', '   TG':"temp", ' Q_TG': 'QTG'}, inplace=True)
```

```python
missing = pd.date_range('1945-04-25', periods=195, freq='D')  # solution for loops from Aron
year44 = pd.date_range('1944-04-25', periods=195, freq='D')
year46 = pd.date_range('1946-04-25', periods=195, freq='D')
m=0
for i in tqdm(missing):
    df.loc[i,"temp"] = (df.loc[year44[m]]["temp"] + df.loc[year46[m]]["temp"]) / 2
    m += 1
```

```python
df['tempK']=df['temp']*0.1+273.15 #Kelvin
df=df.drop(['QTG', 'source','temp'],axis=1) #removing all unnecessary columns
```

```python
df = df[~((df.index.month == 2) & (df.index.day == 29))] #removing all 29 February
```

#### Decomposing

```python
def decompose(df):
   
    df['diff_temp']=df['tempK'].diff() #trend
    df['monthly_mean'] = df.groupby(df.index.month)['diff_temp'].transform('mean')  #seasonality
    df['stationary_temp'] = df.diff_temp - df.monthly_mean  #noise
   
    return df
```

```python
decompose(df).head(2)
```

#### Train/Test Split

```python
train = df.iloc[-3650:-365]['stationary_temp']
test = df.iloc[-365:]['stationary_temp'] 
```

#### ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
train.plot()
```

```python
m = ARIMA(train, order=(1,0,1))
```

```python
m_fit = m.fit()
```

```python
output = m_fit.forecast(steps=365)
```

```python
ypred = output[0]
```

```python
original = df[-365:]
```

```python
original
```

```python
prediction = pd.DataFrame(ypred, index=original.index, columns=['prediction'])
```

```python
plt.figure(figsize=(12,8))
plt.plot(train, color='blue')
plt.plot(test, color='black')
plt.plot(prediction, color='red')
```

#### Reconstructing

```python
pred = pd.concat([prediction, original], axis=1)
```

```python
#1: reseason our data
pred['reseasoned_pred'] = pred.prediction + pred.monthly_mean
```

```python
#2: retrend our data
pred.loc[pred.index[0],'reseasoned_pred'] = pred.loc[pred.index[0],'tempK']
```

```python
# cumulatively sum our data
pred['retrended_data'] = np.cumsum(pred.reseasoned_pred)-273
```

```python
plt.plot(df.tempK[-3650:-365])
plt.plot(pred.index,pred['retrended_data'])
plt.plot(pred.index, df.tempK[-365:])
plt.ylim(250, 310)
```
