```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import seaborn as sns
%matplotlib inline

plt.rcParams['figure.figsize'] = (10,6)
import numpy as np
```

```python
df=pd.read_csv("./data/TG_STAID002759.txt", parse_dates=True, index_col=1)
df.rename(columns = {' SOUID': 'source', '   TG':"temp", ' Q_TG': 'QTG'}, inplace=True)
df['temp']=df['temp']*0.1
```

```python
missing = pd.date_range('1945-04-25', periods=195, freq='D')  # solution for loops from Aron
year44 = pd.date_range('1944-04-25', periods=195, freq='D')
year46 = pd.date_range('1946-04-25', periods=195, freq='D')
m=0
for i in tqdm(missing):
    df.loc[i,"temp"] = (df.loc[year44[m]]["temp"] + df.loc[year46[m]]["temp"]) / 2
    m += 1
```

```python
df['tempK']=df['temp']+273.15 #Kelvin
```

```python
df=df.drop(['QTG', 'source','temp'],axis=1) #removing quality code with 1=suspect and 9=missing
```

### 1. Train/Test Split

```python
train = df.iloc[:41580].copy()  #  from 1876-01-01 to 1989-11-03
test = df.iloc[41580:52800].copy()   #  from 1989-11-04 to 2020-07-23
```

```python
train.plot()
```

### 2. Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
Xtrain = train.index.astype(int).values / 1000_000_000_000_000_000
Xtrain = Xtrain.reshape(-1, 1)
ytrain = train['tempK']
```

```python
m = LinearRegression()
m.fit(Xtrain, ytrain)
m.score(Xtrain, ytrain)  # R^2 (1.0-good, 0.0-bad)
```

```python
Xtest = test.index.astype(int).values / 1000_000_000_000_000_000
Xtest = Xtest.reshape(-1, 1)
ytest = test['tempK']
```

```python
ypred = m.predict(Xtest)
ytrpred = m.predict(Xtrain)
```

### 5. Decomposition
1. Decompose the time series
2. Train a model on the decomposed data
3. Create a prediction from the model
4. Recompose the predictions

```python
train['diff'] = train['tempK'].diff()  # differential
train['pct'] = train['tempK'].pct_change() 
```

```python
train.sample(3)
```

```python
# average daily temperature change:
mean_diff= train['diff'].mean()
```

```python
mean_diff
```

```python
ypred_meandiff = [mean_diff] * 11220
```

```python
mean_pct = train['pct'].mean() + 1
ypred_meanpct = [mean_pct] * 11220
```

```python
# reconstruct to the original scale
# opposite of diff: cumulative sum
ypred_md = np.cumsum(ypred_meandiff) + last
```

```python
# opposite of pct_change: cumulative product
ypred_mp = np.cumprod(ypred_meanpct) * last
```

#### Seasonality

```python
def decompose(train):
    # take out trend + variance change
    train['pct'] = train['tempK'].pct_change()
    mm = train.groupby(train.index.month)['pct'].mean()
    noise = train['pct'] - list(mm) * 3465 # train shape 41580/12 months training data
    return noise, mm
```

```python
noise, mm = decompose(train)
```

```python
# 1. use the monthly mean percentual growth rates
# 2. calculate the cumulative product
def recompose(mm):
    years115 = np.array(list(mm) * 935) + 1 
    result = np.cumprod(years115) * last
    return result
```

```python
ypred_seasperc = recompose(mm)
```

```python
plt.plot(Xtrain, ytrain)
plt.plot(Xtest, ytest)
plt.plot(Xtest, ypred)
plt.plot(Xtrain, ytrpred)

plt.plot(Xtest, ypred_md)   # mean diff
plt.plot(Xtest, ypred_mp)   # mean 
plt.plot(Xtest, ypred_seasperc, 'k--')  # mean seasonal percentage
```

```python
mean_squared_error(ytest, ypred_seasperc)
```
